# Average temperature

```python
# Simulation average-temperature outputs
# These files are written by the corrected C++ programs.

from pathlib import Path


def plot_average_temperature(path, label):
    path = Path(path)
    if not path.exists():
        print(f"Missing file: {path}")
        return
    data = np.loadtxt(path)
    if data.ndim == 1:
        data = data.reshape(1, -1)
    t_min = data[:, 0] / 60.0
    T = data[:, 1]
    if np.nanmean(T) > 100:
        T = T - 273.15
    plt.plot(t_min, T, label=label)

plt.figure(figsize=(9, 5))
plot_average_temperature("/Users/nataliakowalczyk/CLionProjects/Habitat_1/cmake-build-debug/bedroom/average_temperature.dat", "Bedroom simulation")
plot_average_temperature("/Users/nataliakowalczyk/CLionProjects/Habitat_1/cmake-build-debug/geolab/average_temperature.dat", "Geolab simulation")
plt.xlabel("Time [min]")
plt.ylabel("Average temperature [°C]")
plt.title("Average simulated room temperature")
plt.grid(True, alpha=0.3)
plt.legend()
plt.savefig("average_simulated_temperature.png", dpi=200, bbox_inches="tight")
plt.show()
```
